• Solve the following system of Linear equation using Inverse Methods.
2x − 3y + z = −1
x − y + 2z = −3
3x + y − z = 9
{Hint: First use Numpy array to represent the equation in Matrix form. Then Solve for: AX = B}
• Now: solve the above equation using np.linalg.inv function.{Explore more about ”linalg” function
of Numpy}


```python
A = np.array([[3, 4],
              [7, 8]])

B = np.array([[5, 3],
              [2, 1]])

A_inv = np.linalg.inv(A)
identity_check = A.dot(A_inv)
print(identity_check)


AB = A.dot(B)
BA = B.dot(A)

print("AB =\n", AB)
print("BA =\n", BA)

left_side = (A.dot(B)).T
right_side = B.T.dot(A.T)

print(left_side)
print(right_side)

```

```python
A_sys = np.array([[ 2, -3,  1],
                  [ 1, -1,  2],
                  [ 3,  1, -1]])

B_sys = np.array([-1, -3, 9])

A_inv_sys = np.linalg.inv(A_sys)
X_solution = A_inv_sys.dot(B_sys)

print("Solution:", X_solution)
```
