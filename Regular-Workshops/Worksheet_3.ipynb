```python
import pandas as pd

# Load dataset
df = pd.read_csv("/content/drive/MyDrive/aidata/week2/Copy of Titanic-Dataset.csv")

# View basic info
df.info()

# Show first few rows
df.head()
```

```python
# 1. DataFrame with Fare only
fare = df[["Fare"]]
fare.head()
# 2. DataFrame with Pclass and Age
class_age = df[["Pclass", "Age"]]
class_age.head()
# 3. DataFrame with Survived and Sex
survived_gender = df[["Survived", "Sex"]]
survived_gender.head()
```

```python
# 1. Fare > 100
fare_gt_100 = df[df["Fare"] > 100]
fare_gt_100.head()
# 2. Pclass == 1
first_class = df[df["Pclass"] == 1]
first_class.head()
# 3. Age < 18 and Sex = female
female_under_18 = df[(df["Age"] < 18) & (df["Sex"] == "female")]
female_under_18.head()
```

```python
# 1. Embarked in C or S
embarked_c_or_s = df[df["Embarked"].isin(["C", "S"])]
embarked_c_or_s.head()
# 2. Pclass in [1, 2]
first_second_class = df[df["Pclass"].isin([1, 2])]
first_second_class.head()
```

```python
df["Age"] = df["Age"].fillna(df["Age"].median())
# 1. Add fare_per_year column
df["fare_per_year"] = df["Fare"] / df["Age"]

# 2. Subset where fare_per_year > 5
high_fare_age = df[df["fare_per_year"] > 5]

# 3. Sort descending
high_fare_age_srt = high_fare_age.sort_values(by="fare_per_year", ascending=False)

# 4. Select Name + fare_per_year
result = high_fare_age_srt[["Name", "fare_per_year"]]

result.head(10)
```

```python
# 1. Add fare_per_class
df["fare_per_class"] = df["Fare"] / df["Pclass"]

# 2. Filter adult males
adult_males = df[(df["Sex"] == "male") & (df["Age"] >= 18)]

# 3. Sort descending
adult_males_srt = adult_males.sort_values(by="fare_per_class", ascending=False)

# 4. Select Name, Age, fare_per_class
result = adult_males_srt[["Name", "Age", "fare_per_class"]]

result.head(10)
```

```python
# 1. Total fare
total_fare = df["Fare"].sum()

# 2. Fare by each class
fare_class1 = df[df["Pclass"] == 1]["Fare"].sum()
fare_class2 = df[df["Pclass"] == 2]["Fare"].sum()
fare_class3 = df[df["Pclass"] == 3]["Fare"].sum()

# 3. Combine into list
fare_totals = [fare_class1, fare_class2, fare_class3]

# 4. Proportion (percent)
fare_percent = [(x / total_fare) * 100 for x in fare_totals]

fare_percent
```

```python
# 1. Create group column
def age_group(age):
    if age < 18:
        return "child"
    elif age < 65:
        return "adult"
        # age >= 65
    else:
        return "senior"

df["age_group"] = df["Age"].apply(age_group)

# 2. Total passengers
total_passengers = len(df)

# 3. Count per group
group_counts = df["age_group"].value_counts()

# 4. Percent
group_percent = (group_counts / total_passengers) * 100

group_percent
```
